## Feature Engineering

```python
#Adding new feature(integer) for bhk 
df2['BHK']=df2['size'].apply(lambda x: int(x.split(' ')[0]))
df2.BHK.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
        
```

```python
df2[~df2['total_sqft'].apply(is_float)].head(10)
```

```python
def conver_sqft_to_num(x):
    tokens=x.split('-')
    if len(tokens)== 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
conver_sqft_to_num("1320")
```

```python
conver_sqft_to_num("2100 - 2850")
```

```python
df3=df2.copy()
df3['total_sqft']=df3['total_sqft'].apply(conver_sqft_to_num)
df3
```

```python
#Adding feature- price per sqft
df4=df3.copy()
df4['price_per_sqft']=df4['price']*100000/df4['total_sqft']
df4.head(10)
```

```python
#lets check the count of location
len(df4['location'].unique())
```

```python
df4.location= df4.location.apply(lambda x: x.strip())

location_stats= df4.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
#removing outliers
def remove_pps_outlier(df):
    df_out= pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m= np.mean(subdf.price_per_sqft)
        st= np.std(subdf.price_per_sqft)
        reduce_df= subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out= pd.concat([df_out,reduce_df],ignore_index=True)
    return df_out
df6= remove_pps_outlier(df5)
df6.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2= df[(df.location==location) & (df.BHK==2)]
    bhk3= df[(df.location==location) & (df.BHK==3)]
    plt.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price_per_sqft,color='blue',label='2 BHk', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price_per_sqft,marker='*',color='green',label='3 BHk', s=50)
    plt.xlabel('total square feet area')
    plt.ylabel('price per square feet')
    plt.title(location)
    plt.legend()
    
plot_scatter_chart(df6,'Uttarahalli')
```

```python
def remove_bhk_outliers(df):
    exclude_indices=np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df7= remove_bhk_outliers(df6)
df7.shape
```

### Building Model

```python
x=df11.drop('price',axis='columns')
x.head()
```

```python
y= df11.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(x,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'normalize': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_model_using_gridsearchcv(x,y)
```

#### Based on above results we can say that LinearRegression gives the best score.
